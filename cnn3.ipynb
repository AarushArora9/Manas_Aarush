```python
def init_params():
    W1 = np.random.rand(10, 784) - 0.5 # W1=weight of matrix of first layer
    b1 = np.random.rand(10, 1) - 0.5 #b1=nias vector for the first layer
    W2 = np.random.rand(10, 10) - 0.5 #w2=weight matrix for the second layer
    b2 = np.random.rand(10, 1) - 0.5 #b2= bias vector for the second layer
    return W1, b1, W2, b2

def ReLU(Z):
    return np.maximum(Z, 0)

def softmax(Z):
    A = np.exp(Z) / sum(np.exp(Z))
    return A
    
def forward_prop(W1, b1, W2, b2, X):
    Z1 = W1.dot(X) + b1 #X input data matrix, Z1= linear tranformation for layer1
    A1 = ReLU(Z1) #activation of first layer
    Z2 = W2.dot(A1) + b2 #z2=linear transformation for the second layer
    A2 = softmax(Z2)#a2=final out probabilities from the second layer after softmax
    return Z1, A1, Z2, A2

def ReLU_deriv(Z):
    return Z > 0

def one_hot(Y):
    one_hot_Y = np.zeros((Y.size, Y.max() + 1))
    one_hot_Y[np.arange(Y.size), Y] = 1
    one_hot_Y = one_hot_Y.T
    return one_hot_Y

def backward_prop(Z1, A1, Z2, A2, W1, W2, X, Y):
    one_hot_Y = one_hot(Y)
    dZ2 = A2 - one_hot_Y.T#dz2 gradient of the lost wrt to z2, output of second layer
    dW2 = 1 / m * dZ2.dot(A1.T)#gradient of loss wrt to w2
    db2 = 1 / m * np.sum(dZ2)#db2 gradient of the lost with respect to b2
    dZ1 = W2.T.dot(dZ2) * ReLU_deriv(Z1) #dz1=gradient of loss wrt z1, output of first layer
    dW1 = 1 / m * dZ1.dot(X.T)#dw1=gradient of loss wrt w1
    db1 = 1 / m * np.sum(dZ1)#gradient of loss wrt b1
    return dW1, db1, dW2, db2

def update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha):
    W1 = W1 - alpha * dW1
    b1 = b1 - alpha * db1    
    W2 = W2 - alpha * dW2  
    b2 = b2 - alpha * db2    
    return W1, b1, W2, b2
```

```python
def softmax(Z):
    # Subtract the maximum value in Z for numerical stability
    Z_stable = Z - np.max(Z)
    
    # Calculate exponentials
    exp_Z = np.exp(Z_stable)
    
    # Compute the sum of exponentials (adding a small epsilon to avoid division by zero)
    exp_sum = np.sum(exp_Z, axis=0)
    
    # If the sum is zero, replace it with a small value (epsilon)
    exp_sum = np.maximum(exp_sum, 1e-8)  # Avoid division by zero
    
    # Compute softmax
    A = exp_Z / exp_sum
    
    return A
```

```python
import numpy as np

def one_hot(Y):
    # Ensure Y is a numpy array of integers
    Y = np.asarray(Y).astype(int)
    
    # Initialize a matrix of zeros with shape (num_samples, num_classes)
    one_hot_Y = np.zeros((Y.size, Y.max() + 1), dtype=int)
    
    # Set 1 at the positions of the labels
    one_hot_Y[np.arange(Y.size), Y] = 1
    
    return one_hot_Y
```

```python
def get_predictions(A2):
    return np.argmax(A2, 0)

def get_accuracy(predictions, Y):
    print(predictions, Y)
    return np.sum(predictions == Y) / Y.size

def gradient_descent(X, Y, alpha, iterations):
    W1, b1, W2, b2 = init_params()
    for i in range(iterations):
        Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)
        dW1, db1, dW2, db2 = backward_prop(Z1, A1, Z2, A2, W1, W2, X, Y)
        W1, b1, W2, b2 = update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)
        if i % 10 == 0:
            print("Iteration: ", i)
            predictions = get_predictions(A2)
            print(get_accuracy(predictions, Y))
    return W1, b1, W2, b2
```

```python
def make_predictions(X, W1, b1, W2, b2):
    _, _, _, A2 = forward_prop(W1, b1, W2, b2, X)
    predictions = get_predictions(A2)
    return predictions
from matplotlib import pyplot as plt
def test_prediction(index, W1, b1, W2, b2):
    current_image = X_test[:, index, None]
    prediction = make_predictions(X_test[:, index, None], W1, b1, W2, b2)
    label = Y_test[index]
    print("Prediction: ", prediction)
    print("Label: ", label)
    
    current_image = current_image.reshape((28, 28)) * 255
    plt.gray()
    plt.imshow(current_image, interpolation='nearest')
    plt.show()
```
